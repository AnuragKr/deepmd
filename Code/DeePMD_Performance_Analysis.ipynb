```python
import numpy as np
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
from matplotlib import ticker
import seaborn as sns
import math
```

```python
#Specifying file path
base_file_path = "drive/MyDrive/DeepMD_MTP_Project/Code/"
```

```python
df = pd.DataFrame({
    'GPU Device Configuration':['TU102 GPU','V100 GPU','TU102 GPU','V100 GPU','TU102 GPU','V100 GPU','TU102 GPU','V100 GPU'],
    'Cut-Off Radius':['6','6','7','7','8','8','9','9'],
    'Execution Time': [5414,1732,7447,1933,8502,2527,13213,3014]
})

df_1 = pd.DataFrame({
    'Cut-Off Radius':[6,7,8,9],
    'Speedup': [2.12,2.85,2.36,3.38]
})
```

```python
plt.figure(figsize=(8,4),dpi=100)
ax = sns.barplot(x='Cut-Off Radius',y='Execution Time',data=df, hue='GPU Device Configuration', palette="Set2")
ax.set(xlabel='Cut-Off Radius', ylabel='Execution Time',title='Execution Time vs Cut-Off Radius')
ax.set_yscale('log')
plt.legend(title = 'GPU Device', bbox_to_anchor=(1, 1))
plt.tight_layout(pad=1.0)
plt.savefig(base_file_path + 'exec_time_vs_radius.png',bbox_inches = 'tight')
```

```python
plt.figure(figsize=(8,4),dpi=100)
ax = sns.lineplot(x='Cut-Off Radius',y='Speedup',data=df_1, palette="Set1")
ax.set(xlabel='Cut-Off Radius', ylabel='Speedup',title='Speedup vs Cut-Off Radius')
plt.tight_layout(pad=1.0)
plt.savefig(base_file_path + 'speed_up_vs_radius.png',bbox_inches = 'tight')
```

```python
fig, ax1 = plt.subplots(figsize=(12,6))

sns.barplot(x='Cut-Off Radius',y='Execution Time',data=df, hue='GPU Device Configuration', palette="Set2",ax=ax1)
ax2 = ax.twinx()
sns.lineplot(ax=ax2,x=ax.get_xticks(),y='Speedup',data=df_1, palette="Set2",marker='o')
```
