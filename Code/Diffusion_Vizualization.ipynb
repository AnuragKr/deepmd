```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
diffusion = np.loadtxt("diffusion_ice_pimd.txt")
```

```python
Ntraj = diffusion.shape[0]
```

```python
plt.figure(figsize=(12, 8))
plt.title("Mean Squared Displacement", fontsize=12)
plt.xlabel("Time $t$ / ps", fontsize=12)
plt.ylabel("$\langle\\frac{1}{N}\\sum_{i=1}^{N}(\mathbf{r}_i(t)-\mathbf{r}_i(0))^2\\rangle$ / $\mathrm{\AA}^2$", fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.plot(0.05*np.arange(Ntraj), diffusion[:, 1], label="O")
plt.plot(0.05*np.arange(Ntraj), diffusion[:, 2], label="H")
plt.legend(fontsize=12)
```

```python
plt.figure(figsize=(12, 8))
plt.title("Diffusion Coefficient", fontsize=12)
plt.xlabel("Time $t$ / ps", fontsize=12)
plt.ylabel("$D\ / \mathrm{\AA}^2\cdot\mathrm{ps}^{-1}$", fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.plot(diffusion[1:, 0], diffusion[1:, 3], label="O")
plt.plot(diffusion[1:, 0], diffusion[1:, 4], label="H")
plt.legend(fontsize=12)
```

```python
print(diffusion[Ntraj-1, 3])

# Water diffusion coefficient
```
